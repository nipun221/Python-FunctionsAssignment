```python
# 1. What is the difference between a function and a method in Python?

'''
Function: A function is an independent, reusable block of code that is defined
generally and can be called from anywhere in your program by its name. It does
not need to be associated with a specific object or class instance.

Method: A method is a function that is bound to an object. It is defined within
a class and is called on an instance of that class. Methods inherently have
access to the object's data (its attributes) through the special first parameter,
typically named self.
'''

# Examples:

# Functions Examples
def add(a, b):
    return a + b

result = add(5, 3)
print(result)

# Methods Examples
class Dog:
    def __init__(self, name):
        self.name = name

    def bark(self):
        # 'self' allows the method to access the 'name' attribute
        print(f"{self.name} says Woof!")

my_dog = Dog("Fido")
# The method 'bark' is called on the 'my_dog' instance
my_dog.bark()
```

```python
# 2. Explain the concept of function arguments and parameters in Python.

'''
Function arguments and parameters are fundamental concepts in Python used to pass
data to functions during execution.

Parameters: A parameter is a named entity in the function's definition that
specifies the type of information the function expects to receive.
=> In the function def greet(name):, name is the parameter.

Arguments: An argument is the actual value or variable that is passed to
the function when it is called or invoked.
=> When calling greet("Alice"), "Alice" is the argument passed to the name parameter.
'''

# Example:

def add_numbers(a, b): # a, b are parameters
    return a + b

result = add_numbers(5, 10) # 5, 10 are arguments
```

```python
# 3. What are the different ways to define and call a function in Python?

'''
There are mainly three ways to define and call a function in python, each suited
to different programming needs.

A) Standard Functions: This is the most common way to define a named function.
It is a block of reusable code that perform a specific task. It is defined using
the def keyword.

B) Lambda Functions: Lambda functions are small, anonymous functions defined with
the lambda keyword. They can take any number of arguments but can only have one
expression. They are often used when a function is needed for a short period in
conjunction with functions like filter(), map(), and sorted().

C) Functions with Arbitrary Arguments: A function can be defined to accept a
variable number of positional arguments using *args and a variable number of
keyword arguments using **kwargs.

D) Recursive Functions: A function that calls itself to solve a smaller version
of its problem until a base condition is met.

E) Nested Functions: Functions can be defined inside other functions. The inner
function is local to the outer function and can access variables from the outer
function's scope (a concept known as a closure).
'''

# Examples:

# A) Standard Function:
def greet(name):
    """This function greets the person passed in as a parameter."""
    return f"Hello, {name}!"

message = greet("Alice")
print(message)

# B) Lambda Function:
lambda_add = lambda x, y: x + y

add_result = lambda_add(5, 3)
print(add_result)

# C) Function with Arbitrary Arguments:
def my_sum(*args):
    """Sums all positional arguments."""
    return sum(args)

def print_details(**kwargs):
    """Prints key-value pairs."""
    for key, value in kwargs.items():
        print(f"{key}: {value}")

total = my_sum(1, 2, 3, 4)
print(total)

print_details(name="Bob", age=30, city="New York")

# D) Recursive Function:
def factorial(n):
    if n == 1:
        return 1
    else:
        return n * factorial(n - 1)

fac_result = factorial(4)
print(fac_result)

# E) Nested Function:
def outer_function(x):
    def inner_function(y):
        return x + y
    return inner_function

add_five = outer_function(5)
nes_result = add_five(10)
print(nes_result)
```

```python
# 4. What is the purpose of the `return` statement in a Python function?

'''
The return statement in a Python function serves two primary purposes:

A) Exiting the function: When the return statement is executed, the function
immediately terminates, and program control is passed back to the point where
the function was called. Any code following the return statement within the
function will not be executed.

B) Returning a value: The return statement can also pass a value (or values)
back to the caller of the function. This allows functions to perform a calculation
or an action and then provide a result that can be used elsewhere in the program.
'''

# Examples:

def add_numbers(a, b):
    sum_result = a + b
    return sum_result # Returns the calculated sum

def greet(name):
    print(f"Hello, {name}!")
    # No return statement, so it implicitly returns None

# Calling the function with a return value
result = add_numbers(5, 3)
print(f"The sum is: {result}")

# Calling the function with an implicit None return
greeting_result = greet("Alice")
print(f"Greet function returned: {greeting_result}")
```

```python
# 5. What are iterators in Python and how do they differ from iterables?

'''
An iterator is an object that adheres to the iterator protocol. This protocol
requires implementing the __iter__() method, which returns the iterator object
itself, and the __next__() method, which returns the next item from the sequence
or raises StopIteration when there are no more items. Iterators are stateful,
keeping track of their position. They are also memory-efficient as they process
items one by one. You can manually interact with iterators using iter() to
obtain an iterator from an iterable and next() to get the subsequent item.

Iterables and iterators differ in purpose, methods, and behavior.
Feature                         Iterable	                                      Iterator
1. Definition	                   An object that can be looped over.	             An object that manages the iteration state.
2. Key Method(s)	               Implements __iter__() (or __getitem__()).	     Implements both __iter__() and __next__().
3. State	                       Does not maintain iteration state.	             Maintains its state (current position).
4. Relation	                     Not every iterable is an iterator.	             Every iterator is also an iterable.
'''

# Examples:

my_list = [1, 2, 3]           # An iterable
my_iterator = iter(my_list)   # Creates an iterator from the iterable

print(next(my_iterator))
print(next(my_iterator))
print(next(my_iterator))
# next(my_iterator) would raise StopIteration
```

```python
# 6. Explain the concept of generators in Python and how they are defined.

'''
In Python, a generator is a special type of function or expression that allows
you to generate a sequence of values one at a time, making them memory-efficient,
especially for large or infinite sequences. Generators are a powerful way to
implement iterators.

A generator function looks like a regular Python function, but instead of using
the return statement to give back a single value and terminate, it uses the yield
statement.
'''

# Example:

def count_up_to(max_val):
    count = 1
    while count <= max_val:
        yield count  # Pauses and returns the current count
        count += 1

# Using the generator
counter = count_up_to(5)
for number in counter:
    print(number)
```

```python
# 7. What are the advantages of using generators over regular functions?

'''
Generators in Python offer distinct advantages over regular functions when
dealing with large datasets, infinite sequences, or memory-intensive operations.

Key Advantages of Generators:

A) Memory Efficiency (Lazy Evaluation): Regular functions compute and return all
results in a list at once, storing the entire list in memory. Generators only
produce one item at a time using the yield statement and pause their state until
the next value is requested. This makes them ideal for large or infinite data streams.

B) Performance: The "generate-as-needed" approach means you don't wait for all
items to be processed upfront, which improves the initial responsiveness of your program.

C) Handling Infinite Streams: Because generators evaluate lazily, they can represent
infinite sequences (like all prime numbers) without running out of memory, which is
impossible with a standard function returning a list.
'''

# Example:

def squares_list(n):
    result = []
    for i in range(n):
        result.append(i * i)
    return result

# Calling this function with a large number like 1,000,000
# requires a large amount of memory to store all 1 million squares.
# my_list = squares_list(1000000)

def squares_generator(n):
    for i in range(n):
        yield i * i

# This approach uses minimal memory because it only keeps track
# of the *current* number being processed.
for square in squares_generator(1000000):
    # Process each square one by one without storing all of them
    pass
```

```python
# 8. What is a lambda function in Python and when is it typically used?

'''
A lambda function in Python is a small, anonymous function defined without a
name using the lambda keyword. It can take any number of arguments but can only
have one expression.

Syntax => lambda arguments: expression

Common use cases include:

A) filter(): To filter elements from an iterable based on a condition.

B) map(): To apply an operation to every element in an iterable.

C) sorted(): To define a custom sort key for sorting a list of complex objects
(e.g., tuples, dictionaries).
'''

# Examples:

# A) filter(): Filter out even numbers
my_list = [1, 2, 3, 4, 5]
filtered_list = list(filter(lambda x: x % 2 != 0, my_list))
# Result: [1, 3, 5]

# B) map(): Square each number
my_list = [1, 2, 3, 4, 5]
squared_list = list(map(lambda x: x * x, my_list))
# Result: [1, 4, 9, 16, 25]

# C) sorted(): Sort a list of tuples by the second element
data = [('apple', 10), ('banana', 3), ('cherry', 8)]
sorted_data = sorted(data, key=lambda x: x[1])
# Result: [('banana', 3), ('cherry', 8), ('apple', 10)]
```

```python
# 9. Explain the purpose and usage of the `map()` function in Python.

'''
The primary purpose of map() is to provide a concise and efficient way to perform
a transformation on every element of an iterable without using an explicit for loop.
This functional programming approach is often cleaner and, for certain operations,
more memory-efficient because it returns an iterator that generates results on the
fly (lazily) rather than creating a new list with all the results immediately.

The syntax for the map() function is => map(function, iterable, ...)

It accepts two required arguments:

A) function: The function to which each item of the iterable will be passed.

B) iterable: The sequence (list, tuple, etc.) whose elements are transformed.
Multiple iterables can be provided if the function accepts multiple arguments.
'''

# Examples:

# Using a named function
def square_number(n):
    return n * n

numbers = [1, 2, 3, 4, 5]

# Map the square_number function to the numbers list
squared_map = map(square_number, numbers)

# Convert the map object to a list to view the results
squared_list = list(squared_map)

print(squared_list)
# Output: [1, 4, 9, 16, 25]


# Using a lambda function
numbers = [10, 20, 30, 40]

# Use a lambda to multiply each number by 2
doubled_list = list(map(lambda x: x * 2, numbers))

print(doubled_list)
# Output: [20, 40, 60, 80]
```

```python
# 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

'''
The core difference between Python's built-in map(), filter(), and reduce()
functions lies in how they apply a given function to an iterable and what they return.

A) map() transforms each element of an iterable into a new element based on a
function and returns a new iterable of the transformed values.

B) filter() selects elements from an iterable for which a function returns True
and returns a new iterable containing only the selected elements.

C) reduce() applies a function cumulatively to the items of an iterable to
produce a single, aggregated result value.
'''

# Examples:

# A) map()
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
# Map the square function over the numbers
squared_numbers_map = map(square, numbers)

# Convert to a list to view the result
print(list(squared_numbers_map))
# Output: [1, 4, 9, 16]

# B) filter()
def is_even(x):
    return x % 2 == 0

numbers = [1, 2, 3, 4, 5, 6]
# Filter the numbers using the is_even condition
even_numbers_filter = filter(is_even, numbers)

# Convert to a list to view the result
print(list(even_numbers_filter))
# Output: [2, 4, 6]

# C) reduce()
from functools import reduce

def add(x, y):
    return x + y

numbers = [1, 2, 3, 4]
# Reduce the list by repeatedly applying the add function
total = reduce(add, numbers)

print(total)
# Output: 10
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms
        self.current_term = 0
        self.a = 0
        self.b = 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.terms:
            raise StopIteration

        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            next_term = self.a + self.b
            self.a, self.b = self.b, next_term
            self.current_term += 1
            return next_term

fib_seq = FibonacciIterator(10)

print("Fibonacci sequence up to 10 terms:")
for num in fib_seq:
    print(num)

fib_seq_5 = FibonacciIterator(5)
print("\nFibonacci sequence up to 5 terms:")
print(next(fib_seq_5))
print(next(fib_seq_5))
print(next(fib_seq_5))
print(next(fib_seq_5))
print(next(fib_seq_5))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def power_of_two(exponent):
  for i in range(exponent + 1):
    yield 2**i

for power in power_of_two(5):
  print(power)
```

```python
# 10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
  vowels = 'aeiouAEIOU'
  result_list = filter(lambda char: char not in vowels, input_string)
  result_string = "".join(result_list)
  return result_string

string = "Hello, World!"
result = remove_vowels(string)
print(result)
```
